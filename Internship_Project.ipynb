```python
import pandas as pd
```

```python
cancer = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv')
```

```python
cancer.head()
```

```python
cancer.info()
```

```python
cancer.describe()
```

```python
# Define target (y) and features (X)
cancer.columns
```

```python
y = cancer['diagnosis']
X = cancer.drop(['id','diagnosis','Unnamed: 32'],axis=1)
```

```python
# Train-Test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# Check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# select model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
```

```python
# Fit model
model.fit(X_train,y_train)
LogisticRegression(max_iter=5000)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Predict model
y_pred = model.predict(X_test)
y_pred
```

```python
# Model accuracy
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# Confusion Matrix Heatmap
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Classification Report Bar Plot
report = classification_report(y_test, y_pred, output_dict=True)
df_report = pd.DataFrame(report).transpose()

metrics = ['precision', 'recall', 'f1-score']
classes = ['B', 'M']

fig, ax = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

for i, metric in enumerate(metrics):
    sns.barplot(x=classes, y=[df_report.loc[cls, metric] for cls in classes], ax=ax[i], palette='viridis')
    ax[i].set_title(f'{metric.capitalize()} by Class')
    ax[i].set_xlabel('Class')
    ax[i].set_ylabel(metric.capitalize())
    ax[i].set_ylim(0, 1)

plt.suptitle('Classification Report Metrics', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```
